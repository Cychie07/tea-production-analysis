# Tea Oxidation Analysis

This notebook provides a simple analysis of enzymatic oxidation data for black tea production on January 16, 2025.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Excel
df = pd.read_excel("tea_production_simulated.xlsx")
df.head()
```

## Average Oxidation Time per Bed


```python
beds = ["Lama Oksidasi 1 (menit)", "Lama Oksidasi 2 (menit)",
        "Lama Oksidasi 3 (menit)", "Lama Oksidasi 4 (menit)"]
df["Rata-rata Oksidasi"] = df[beds].mean(axis=1)

plt.figure(figsize=(10, 5))
plt.plot(df["Jam"], df["Rata-rata Oksidasi"], marker='o', color='darkgreen')
plt.title("Rata-rata Durasi Oksidasi Enzimatis per Jam")
plt.xlabel("Jam Produksi")
plt.ylabel("Durasi Oksidasi (menit)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

##  Insight
- The average enzymatic oxidation time ranges from 100 to 120 minutes.
- Time blocks exceeding 115 minutes tend to produce dull-colored tea with a pungent flavor.

- The optimal oxidation duration is estimated to be in the range of 105–112 minutes for balanced quality.
